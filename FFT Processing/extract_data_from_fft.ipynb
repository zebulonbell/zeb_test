```python
import pandas as pd
import os
```

```python
df_final = pd.DataFrame(columns = ['path', 'axis', 'date', '24.854368932038835', '29.825242718446603', '34.79611650485437',
                                   '54.679611650485434', '59.650485436893206', '64.62135922330097'])

results_path = r"D:\Projects\GP QRCM\fft"

os.chdir(results_path)
current_path = os.getcwd()
print(f"Current Path is set to {current_path}")


for root, dirs, files in os.walk(results_path):
    #print(root)
    for file in files:
        file_path = os.path.join(root, file)
        print(file_path)
        
        df = pd.read_csv(file_path)
    
        df_max_values = df.groupby(['path', 'axis', 'date']).agg({'24.854368932038835' : 'max', 
                                '29.825242718446603'  : 'max', 
                                '34.79611650485437'  : 'max', 
                                '54.679611650485434' : 'max', 
                                '59.650485436893206'  : 'max', 
                                '64.62135922330097' : 'max'}).reset_index()

        print(df_max_values.shape)
        
        df_final = df_final.append(df_max_values, ignore_index = True)
        #print(df_final.shape)       

        

df_final['Max1'] = df_final.apply(lambda x: max(x['24.854368932038835'], x['29.825242718446603'], x['34.79611650485437']), axis =1)
df_final['Max2'] = df_final.apply(lambda x: max(x['54.679611650485434'], x['59.650485436893206'], x['64.62135922330097']), axis =1)

levels = df_final['path']
df_final[['L1', 'L2', 'L3', 'L4', 'L5']] = levels.str.split("/", n = 5, expand = True) 

#df_final
output_file_path = os.path.join("D:\Projects\GP QRCM\output", 'fft_out.csv')
df_final.to_csv(output_file_path, index = None)
```
